Remeber: A prime number (or a prime) is a natural number greater than 1 that has no positive divisors other than 1 and itself. A natural number greater than 1 that is not a prime number is called a composite number.

```python

    prime = False
    while not prime:
        print(f"-- check if {p} is prime:")
        for k in range(2, p//2 + 1):
            if p % k == 0:
                print(f" no! Divides by {k}")
                break
        else:
            print(" yep!!!")
            prime = True
        p += 1
```

````python
import math
n = 1000
p = 2
primes = [2]

for i in range(3, n, 2): # only test odd numbers, as only 2 is even prime
    print(f"\n Testing {i}...", end="")
    for k in primes:
        if i % k == 0:
            print(f" it's not a prime! Divides by {k}")
            break
        elif k > math.sqrt(i): # no need to test further
            print(" yep!!!")
            primes.append(i)
            break
print(primes)
````

Create a program that gets the marks the user has for their classes this year and tells them how many classes they are failing.

````python
classes = ["Math", "English", "Science", "History", "French"]
marks = {}
for c in classes:
    marks[c] = float(input(f"What is your mark in {c}? "))

failed = filter(lambda x: x < 10, marks.values())
print("You are failing", len(list(failed)), "classes")
````

Write a program to be used by student council in this upcoming election to tabulate the votes after the election. Your program will ask the number of candidates and their names. Then, the user will enter all of the ballets (1 for the first candidate, 2 for the second candidate, ...) until (s)he enter 0
The program will tell who won the election and what percentage each candidate earned.

```
    Graham Chapman
    John Cleese
    Terry Gilliam
    Enter Vote (0 to exit) : 2
    Enter Vote (0 to exit) : 2
    Enter Vote (0 to exit) : 2
    Enter Vote (0 to exit) : 1
    John Cleese won
    John Cleese earned 75\% of the votes
    Graham Chapman earned 25\% of the votes
    Terry Gilliam earned 0\% of the votes
```

````python
n = int(input("How many candidates? "))
candidates = []

# get candidates names
for i in range(n):
    candidates.append(input(f"Enter name of candidate {i+1}: "))

# get votes
votes = []
while True:
    vote = int(input("Enter vote (0 to exit): "))
    if vote == 0:
        break
    votes.append(vote)

# count votes
counts = [0] * n
for v in votes:
    counts[v-1] += 1

# print results
for i in range(n):
    print(f"{candidates[i]} earned {counts[i]/len(votes)*100:.0f}% of the votes")
````

Palindromes are words that are the same forwards and backwards.  Create a program that gets a word from the user and tells them if it is palindrome or not.  Your program should not care about capitalization. For instance, test with the following:
- rotator (should be a palindrome)
- racecar (should be a palindrome)
- kayak (should be a palindrome)
- python (should not be a palindrome)

````python
word = input("Enter a word: ")
if word.lower() == word[::-1].lower():
    print("It's a palindrome!")
else:
    print("It's not a palindrome!")
````
